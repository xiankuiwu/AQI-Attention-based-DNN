```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.family'] = 'Times New Roman'
```

```python
df = pd.read_csv('beijing.csv',encoding = 'gb2312')
```

```python
df.head()
```

```python
time=df.iloc[0:200,0]
```

```python
print(time)
```

```python
aqi_data = df.iloc[0:200,1]
```

```python
print(aqi_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt


time_numeric = np.arange(len(time)) + 1
data = np.column_stack((time_numeric, aqi_data))
alpha = 0.05

def exponential_smoothing(alpha, s):
    s_hat = np.zeros_like(s)
    s_hat[0] = s[0]
    for t in range(1, len(s)):
        s_hat[t] = alpha * s[t] + (1 - alpha) * s_hat[t - 1]
    return s_hat

smoothed_aqi_data = exponential_smoothing(alpha, data[:, 1])


plt.figure(figsize=(6, 4))
plt.plot(data[:, 0], data[:, 1], label='AQI')
plt.plot(data[:, 0], smoothed_aqi_data, label=f'AQI_ETS (alpha={alpha})', linestyle='dashed')


plt.text(1, max(data[:, 1]), '#1', fontsize=12, verticalalignment='top', horizontalalignment='left')

plt.xlabel('Time')
plt.ylabel('AQI')
plt.legend()
plt.savefig('BA.pdf', bbox_inches='tight')
plt.show()
```
